```python
#A moving average (MA) is a widely used indicator in technical analysis that helps smooth out
#price action by filtering out the “noise” from random short-term price fluctuations. 
#It is a trend-following, or lagging, indicator because it is based on past prices.
ma_day = [10,20,50]

for ma in ma_day:
    column_name = "MA for %s days" % (str(ma))
    AAPL[column_name] = pd.DataFrame.rolling(AAPL['Close'], ma).mean()
    
AAPL[['Close','MA for 10 days','MA for 20 days','MA for 50 days']].plot(subplots=False,figsize=(10,4))
```

```python
# A Stock Heat Map is a two-dimensional visual Chart,
#representing the performance of stocks listed on stock exchanges, useful for traders and investors.
#Good stock screener tools will allow you to set different criteria and the heat map will change based on your settings.
#It is very quick because of the different colors and size of the tiles - the visualization of the stock market is synoptical.
#Compute the correlation matrix
corr = stocks_return.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
